```python
import pandas as pd 
import re 
import datetime
import random

columns = [
            "changenumber",
            "taknumber",
            "requested_by",
            "configuration_item",
            "risk",
            "impact",
            "assignment_group",
            "assigned_to"
            "state",
            "type"
            "shortdesc",
            "implementation_start_date",
            "implementation_end_date"
        ]
```

```python
def generate_alpha_numeric_ids(sample , n):
    
    split_chars = re.split('(\d+)',sample)
    alpha = split_chars[0]
    task_list = []
    for i in range(n):
        task_list.append(alpha+str(int(split_chars[1])+i))
        
    return task_list
```

```python
def generate_from_list(sample,n):
    op_list = []
    for i in range(n):
        e = random.choice(sample)
        op_list.append(e)
    
    return op_list 
```

```python

def generate_date_range(start , end , n):
    
    return pd.date_range(start, end, periods=n)
```

```python
requested_by_list=["Yogesh","Suresh","Divya","Dixit","Sharmila"]
configuration_item_list=["BAFIR-PARIS","DB-SIS","EBICS,TSS-CDS Workflow","FMS-EMEA"]
impact_list=["high", "low"," medium"]
urgency_list=["high", "low"," medium"]
assignment_group_list=["GDR_L2_Support","CTAS_L2_Support","TF3_L2_Support","MESH_L3-SUPPORT","UNIX_BESTSHORE-SUPPORT"]
assigned_to_list=["Arpita","Livingston","Priya","Franklin","Richny"]
implemented_state_list=["Implementation","Closed","Closed","Implementation","Approval"]
type_list=["Normal","Standard"]
short_description_list=["Add user","Read only access to database","Account revoke","Create Mesh18 access","Delete user"]
random_days = [5,6,7,8,9,10]

def changerequest_dataset(n,cols):
    
        chang_no=generate_alpha_numeric_ids("CHG000121314" , n)
        task_no=generate_alpha_numeric_ids("TASK0006775904" , n)
        requested_by=generate_from_list(requested_by_list,n)
        configuration_item=generate_from_list(configuration_item_list,n)
        impact=generate_from_list(impact_list,n)
        urgency=generate_from_list(impact_list,n)
        assignment_group=generate_from_list(assignment_group_list,n)
        assigned_to=generate_from_list(assigned_to_list,n)
        implemented_state=generate_from_list(implemented_state_list,n)
        type1=generate_from_list(type_list,n)
        short_description=generate_from_list(short_description_list,n)
        opendon=generate_date_range("2020-08-01" , "2021-02-05" , n)


        
        generated_df = pd.DataFrame(columns=cols)
        generated_df["changenumber"]=chang_no
        generated_df["taknumber"]=task_no
        generated_df["requested_by"]=requested_by
        generated_df["configuration_item"]=configuration_item
        generated_df["risk"]=impact
        generated_df["impact"]=impact
        generated_df["assignment_group"]=assignment_group
        generated_df["assigned_to"]=assigned_to
        generated_df["state"]=implemented_state
        generated_df["type"]=type1
        generated_df["shortdesc"]=short_description
        generated_df["implementation_start_date"]=opendon
        end_time_list = []
        start_time_list = []
        for i in generated_df["implementation_start_date"]:

            e = random.choice(random_days)
            d=i+datetime.timedelta(days=e)
            end_time_list.append(d.date())
            start_time_list.append(i.date())

        generated_df["opened_on"] = start_time_list
        generated_df["implementation_end_date"] = end_time_list
        generated_df.to_csv("change_Data.csv")
        return generated_df
       
```
