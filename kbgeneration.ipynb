```python
import pandas as pd 
import random 
import re
import datetime
columns = [
           "kb_number",
            "configuration_item",
            "opened_on",
            "opened_by",
            "workflow",
            "asisgnment_group",
            "valid_to",
            "short_description"
          ]
```

```python
def generate_alpha_numeric_ids(sample , n):
    
    split_chars = re.split('(\d+)',sample)
    alpha = split_chars[0]
    kb_list = []
    for i in range(n):
        kb_list.append(alpha+str(int(split_chars[1])+i))
        
    return kb_list
```

```python
def generate_from_list(sample,n):
    op_list = []
    for i in range(n):
        e = random.choice(sample)
        op_list.append(e)
    
    return op_list 
```

```python

def generate_date_range(start , end , n):
    
    return pd.date_range(start, end, periods=n)
```

```python
configuration_item_list=["BAFIR-PARIS","DB-WEB Investor","DB-SIS","DB-TF3","DB-CORAL"]
opened_by_list=["Abhy","Suresh","Dhurka","Vignesh","Rakesh"]
workflow_list=["Draft","Expired"]
assignment_group_list=["BAFIR_L2_SUPPORT","CTAS_L2_SUPPORT","CTAS_L2_SUPPORT","TSS_TF3_SUPPORT","TRADE_L2_SUPPORT"]
short_description_list=["RSS Report Generation","Netprobe is down","Account Unlock","User unable to login"]
random_days = [5,6,7,8,9,10]

def kbrequest_dataset(n,cols):
    kb_no=generate_alpha_numeric_ids("KB00146" , n)
    configuration_item=generate_from_list(configuration_item_list,n)
    opendon=generate_date_range("2020-08-01" , "2021-02-05" , n)
    opened_by=generate_from_list(opened_by_list,n)
    workflow=generate_from_list(workflow_list,n)
    assignment_group=generate_from_list(assignment_group_list,n)
    short_description=generate_from_list(short_description_list,n)
    
    generated_df = pd.DataFrame(columns=cols)
    generated_df["kb_number"]=kb_no
    generated_df["configuration_item"]=configuration_item
    generated_df["opened_by"]=opened_by
    generated_df["workflow"]=workflow
    generated_df["asisgnment_group"]=assignment_group
    generated_df["short_description"]=short_description
    generated_df["opened_on"]=opendon
        
    end_time_list = []
    start_time_list = []
    for i in generated_df["opened_on"]:

            e = random.choice(random_days)
            d=i+datetime.timedelta(days=e)
            end_time_list.append(d.date())
            start_time_list.append(i.date())

    generated_df["opened_on"] = start_time_list
    generated_df["valid_to"] = end_time_list
    generated_df.to_csv("kb_Data.csv")
    return generated_df
```

```python
kbrequest_dataset(1000,columns)
```
