```python
import pandas as pd 
import re
import random
import datetime
```

```python
columns = [
           "request_item",
           "Task_Reference_Number",
           "configuration_item",
           "assignment_group",
           "assigned_to",
           "request_type",
           "request_subtype",
           "priority",
           "state",
           "urgency",
           "short_description",
          "description",
          "opened_by",
          "requested_for",
          "opened_on",
          "due_date"

    ]
```

```python
def generate_alpha_numeric_ids(sample , n):
    
    split_chars = re.split('(\d+)',sample)
    alpha = split_chars[0]
    task_list = []
    for i in range(n):
        task_list.append(alpha+str(int(split_chars[1])+i))
        
    return task_list

#generate_alpha_numeric_ids("RITM023251" , 20) #requestitem
#generate_alpha_numeric_ids("TASK0006775904" , 20)   #taskid
```

```python
def generate_from_list(sample,n):
    op_list = []
    for i in range(n):
        e = random.choice(sample)
        op_list.append(e)
    
    return op_list 
```

```python

def generate_date_range(start , end , n):
    
    return pd.date_range(start, end, periods=n)
```

```python
sr_desc_name = {
                      "Extraction of Report" : "Extract the weekly DB-SIS report",
                       "Account is locked" : "Account is locked. Please unlock it",
                      "Create user account" : "Create an account to the user"
                }
reques_for_opendfor={
                      "Mithran":"Mithran",
                       "Kamal":"Kamal",
                       "vimal":"Vimal"
                     }
def generate_sr_desc_name(sample , n):
    
    srdesc = list(sample.keys())
    
    srdesc_op_list = []
    detailedsrdesc_op_list = []
    
    for i in range(n):
        e = random.choice(srdesc)
        srdesc_op_list.append(e)
        
    # Now mapping the names
        
    for i in srdesc_op_list:
        detailedsrdesc_op_list.append(sample[i])
        
    return srdesc_op_list,detailedsrdesc_op_list
#generate_sr_desc_name(sr_desc_name, 3)
```

```python
configuration_item_list=["DB-SIS","FMS-EMEA","DB-Web Investor"]
assignment_group_list=["CTAS_L2_SUPPORT","FMS_L2_SUPPORT","GDS_L2 SUPPORT"]
assigned_to_list=["Lokesh","John","Vignesh"]
request_type_list=["Report Generation","Account Unlock","Account Maintenance"]
request_subtype_list=["Report","Account Maintenance","Create"]
priority_list=["low","high","medium"]
state_list=["Solved Proposed" , "Solved Confirmed" , "Inprogress Investigate"]
urgency_list=["low","high","medium"]
random_days = [5,6,7,8,9,10]

# short_description_list=["Extraction of Report","Account is locked","Create user account"]
# description_list=["Extract the weekly DB-SIS report","Account is locked. Please unlock it","Create an account to the user"]
#opened_by_list=["Mithran","Kamal","Vimal"]
#requested_for_list=["Mithran","Kamal","Vimal"]

def srrequest_dataset(n,cols):

        request_item=generate_alpha_numeric_ids("RITM023251" , n)
        Task_Reference_Number=generate_alpha_numeric_ids("TASK0006775904" , n)
        configuration_item=generate_from_list(configuration_item_list,n)
        assignment_group=generate_from_list(assignment_group_list,n)
        assigned_to=generate_from_list(assigned_to_list,n)
        request_type=generate_from_list(request_type_list,n)
        request_subtype=generate_from_list(request_subtype_list,n)
        priority=generate_from_list(priority_list,n)
        state=generate_from_list(state_list,n)
        urgency=generate_from_list(urgency_list,n)
        short_description,description=generate_sr_desc_name(sr_desc_name,n)
        opened_by,reques_for=generate_sr_desc_name(reques_for_opendfor,n)
        opendon=generate_date_range("2020-08-01" , "2021-02-05" , n)

        generated_df = pd.DataFrame(columns=cols)
        generated_df["request_item"]=request_item
        generated_df["Task_Reference_Number"]=Task_Reference_Number
        generated_df["configuration_item"]=configuration_item
        generated_df["assignment_group"]=assignment_group
        generated_df["assigned_to"]=assigned_to
        generated_df["request_type"]=request_type
        generated_df["request_subtype"]=request_subtype
        generated_df["priority"]=priority
        generated_df["state"]=state
        generated_df["urgency"]=urgency
        generated_df["short_description"]=short_description
        generated_df["description"]=description
        generated_df["opened_by"]=opened_by
        generated_df["requested_for"]=reques_for
        generated_df["opened_on"]=opendon
        
        end_time_list = []
        start_time_list = []
        for i in generated_df["opened_on"]:

            e = random.choice(random_days)
            d=i+datetime.timedelta(days=e)
            end_time_list.append(d.date())
            start_time_list.append(i.date())

        generated_df["opened_on"] = start_time_list
        generated_df["due_date"] = end_time_list
        generated_df.to_csv("service_Data.csv")
        return generated_df
```
